```python
logger = get_logger()
```

```python
def main_pipeline():
    try:
        try:
            main_ingestion()
            logger.info("Ingestion step executed successfully.")
        except Exception as e:
            logger.error(f"Ingestion step failed: {e}")
            raise

        try:
            main_transformation()
            logger.info("Transformation step executed successfully.")
        except Exception as e:
            logger.error(f"Transformation step failed: {e}")
            raise

        try:
            main_enrich()
            logger.info("Enrichment step executed successfully.")
        except Exception as e:
            logger.error(f"Enrichment step failed: {e}")
            raise

        logger.info("Pipeline executed successfully.")
    except Exception as e:
        logger.error(f"Pipeline execution failed: {e}")
        raise

if __name__ == "__main__":
  main_pipeline()
```

**_Analytical **_Views_**_**

```python
%sql
-- Profit by Year
SELECT year, ROUND(SUM(total_profit), 2) AS profit 
FROM az_ci_de_dbs.ecom_dev.profit_by_year 
GROUP BY year 
ORDER BY year
```

```python
%sql
-- Profit by Year + Product Category
SELECT year, category, ROUND(SUM(total_profit), 2) AS profit 
FROM az_ci_de_dbs.ecom_dev.profit_by_year 
GROUP BY year, category 
ORDER BY year, category
```

```python
%sql
-- Profit by Customer
SELECT customer_name AS customer, ROUND(SUM(total_profit), 2) AS profit 
FROM az_ci_de_dbs.ecom_dev.profit_by_year 
GROUP BY customer_name 
ORDER BY customer_name
```

```python
%sql
-- Profit by Customer + Year
SELECT year, customer_name AS customer, ROUND(SUM(total_profit), 2) AS profit 
FROM az_ci_de_dbs.ecom_dev.profit_by_year 
GROUP BY year, customer_name 
ORDER BY year, customer_name
```
