```python
logger = get_logger()
```

```python
def main_pipeline():
    try:
        try:
            test_utils_main()
            logger.info("Test utils executed successfully.")
        except Exception as e:
            logger.error(f"Test utils failed: {e}")
            raise
            
        try:
            main_ingestion()
            logger.info("Ingestion step executed successfully.")
        except Exception as e:
            logger.error(f"Ingestion step failed: {e}")
            raise

        try:
            test_ingestion_main()
            logger.info("Ingestion tests executed successfully.")
        except Exception as e:
            logger.error(f"Ingestion tests failed: {e}")
            raise

        try:
            main_transformation()
            logger.info("Transformation step executed successfully.")
        except Exception as e:
            logger.error(f"Transformation step failed: {e}")
            raise

        try:
            test_transformation_main()
            logger.info("Transformation tests executed successfully.")
        except Exception as e:
            logger.error(f"Transformation tests failed: {e}")
            raise

        try:
            main_enrich()
            logger.info("Enrichment step executed successfully.")
        except Exception as e:
            logger.error(f"Enrichment step failed: {e}")
            raise

        try:
            test_enrich_main()
            logger.info("Enrichment tests executed successfully.")
        except Exception as e:
            logger.error(f"Enrichment tests failed: {e}")
            raise

        logger.info("Pipeline executed successfully.")
    except Exception as e:
        logger.error(f"Pipeline execution failed: {e}")
        raise

if __name__ == "__main__":
  main_pipeline()
```
